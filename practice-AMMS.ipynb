To Determine the terminal velocity of the falling parachutist near earth's surface.
we knows that 
$$F=ma$$ 
 or 
  $$a = {F \over m}$$

A model for this case can be derived by expressing the acceleration as the time rate of change of the veloc足ity $(dv/dt)$ and substituting it to yield 

$${dv \over dt} = {F \over m}$$

where $v$ is velocity (m/s) and t is time (s). Thus, the mass multiplied by the rate of change of the velocity is equal to the net force acting on tbe body. If the net force is positive, the object will accelerate. If it is negative, the object will decelerate. If the net force is zero, the object's velocity will remain at a constant level.

$$F= F_D + F_U$$
If the downward force is assigned a positive sign, the second law can be used to for足mulate the force due to gravity, as
$$F_D = mg$$
where g = the gravitational constant, or the acceleration due to gravity, which is approxi足mately equal to 9.8 m/s2.

Air resistance can be formulated in a variety of ways. A simple approach is to assume that it is linearly proportional to velocity1 and acts in an upward direction, as in 
$$F_U = -cv$$

where $c = a$ proportionality constant called the drag coefficient (kg/s). Thus, the greater the fall velocity, the greater the upward force due to air resistance.

The net force is the difference between the downward and upward force. Therefore,

$${dv \over dt} = {mg-cv \over m}$$
or
$${dv \over dt} = g- {c \over m}v$$

This is a equation is a model of falling body that relates the acceleration of a falling object to the forces act足ing on it. It is a differential equation because it is written in terms of the differential rate of change (dv/ dt) of the variable that we are interested in predicting.

$$v(t) = {gm \over c }(1-e^{-(c/m)t})$$

```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
g = 9.81 #m/s^2 gravity
m = 68.1 #Kg  mass
c = 12.5 #Kg/s drag cofficient
t_0 = 0 #s starting time
N = 50 #s ending time
# delta_t = 1 #steps
v1 = np.zeros(N) #m/s Terminal velocity
t  = range(t_0,N)

m2 = 90 # Chaging mass
c2 =5; #Chaging drag cofficient2
c3 =100; #Chaging drag cofficient3
v2 = np.zeros(N) #Changing Terminal Velocity when m=m2
v3 = np.zeros(N) #Chang in Terminal Velocity3 when c=c2 and m =m
v4 = np.zeros(N) #Chang inTerminal Velocity3 when c=c3 and m = m


for i in t:
    v1[i] = (g*m)/c * (1 - math.exp(-(c/m)*t[i]))
    v2[i] = (g*m2)/c * (1 - math.exp(-(c/m2)*t[i]))
    v3[i] = (g*m)/c2 * (1 - math.exp(-(c2/m)*t[i]))
    v4[i] = (g*m)/c3 * (1 - math.exp(-(c3/m)*t[i]))

    # print("i: {} Velocity1: {}".format(i,v1[i]))
    # print("i: {} Velocity2: {}".format(i,v2[i]))
    # print("i: {} Velocity3: {}".format(i,v3[i]))
    # print("i: {} Velocity4: {}".format(i,v4[i]))


plt.title("Falling Parachutist Velocity Over Time Diff. Mass Comparasion")
plt.plot(t,v1, color="r", linewidth=3, )
plt.plot(t,v2, color="g", linewidth=3)
plt.legend(["mass= 68.1 Kg", "mass = 90 Kg","c = 60"])

plt.xlabel("Time [s]")
plt.ylabel("Velocity [m/s]")
plt.grid()
plt.show()

plt.title("Falling Parachutist Velocity Over Time Diff. Drag Coff. Comparasion")

plt.plot(t,v3, color="r", linewidth=3, )
plt.plot(t,v4, color="g", linewidth=3)
plt.legend(["c= 5 Kg/s", "c = 100 Kg/s"])

plt.xlabel("Time [s]")
plt.ylabel("Velocity [m/s]")
plt.grid()
plt.show()
```
